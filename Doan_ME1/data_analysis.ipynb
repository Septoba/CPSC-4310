## Advertising data

Let's take a look at some data, ask some questions about that data, and then use linear regression to answer those questions!

A toy dataset for your exploratory data analysis:

./data/Advertising.csv

```python
import pandas as pd

# read CSV file stored in the current directory and save the results
data = pd.read_csv('./data/Advertising.csv', skipinitialspace=True, index_col=0)

# display the first 5 rows
data.head()
```

What are the features?
- **TV:** advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
- **Radio:** advertising dollars spent on Radio
- **Newspaper:** advertising dollars spent on Newspaper

What is the response?
- **Sales:** sales of a single product in a given market (in thousands of items)

What else do we know?
- Because the response variable is continuous, this is a **regression** problem.
- There are 200 **observations** (represented by the rows), and each observation is a single market.

```python
import numpy as np
import matplotlib.pyplot as plt
print(data.agg([np.mean, np.median]))
print(data.mode())
```

```python
print(data.agg([np.min, np.median, np.max]))
print(data.quantile([.25,.75]))
```

```python
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [10, 5]
plt.boxplot(data)
plt.show
```

```python
plt.hist(data.Sales)
plt.show
print('The histogram of the sales data is unimodal with a slightly positive skew')
```

```python
plt.plot(data.Sales)
plt.show
```

```python
data.corr(method='pearson')
```

```python
plt.scatter(data.TV, data.Sales)
plt.scatter(data.Radio, data.Sales, marker='>')
plt.scatter(data.Newspaper, data.Sales, marker='s')
plt.legend(data)
```

While all three graphs above uses the same data to generate those graphs, they have different interpretations in how each could be read. The boxplot graphs the five point summary for each resources in the data set, the histogram counts how much sales there is in the data set, and lastly, the scatter plot plops all of the sales into a graph to see if there is a pattern or relationship bewteen those variables.
